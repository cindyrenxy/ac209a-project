# 1. Problem Overview

Predicting patient outcomes in the intensive care unit (ICU) is a critical challenge. Early identification of patients at high risk of mortality enables clinicians to prioritize interventions, allocate resources efficiently, and improve overall patient care. Machine learning and statistical modeling approaches can serve as powerful tools for improving outcome prediction as they can capture complex relationships and patterns inherent in patient data. 

However, due to the high dimensionality, missingness, and correlations present in ICU datasets, it is critical to develop models that balance predictive performance with interpretability and clinical practicality. Logistic regression models offer simple and interpretable quantifications of the associations between predictors and outcomes, whereas random forests can model complex, nonlinear relationships and interactions between variables.

By applying both of these methods to ICU data collected within the first 24 hours of admission, this project aims to identify key predictors of patient mortality and construct a predictive model. Through evaluation metrics such as area under the curve (AUC) and precision and recall, we will demonstrate the feasibility of implementing our model in real-world ICU settings to support timely and evidence-based decision making.

How can machine learning models be used to predict the risk of patient mortality in the intensive care unit (ICU) within the first 24 hours of admission?

## 5.1 Feature Selection

In this section, we explored feature selection strategies to improve model stability, reduce noise, and streamline the downstream predictive workflow. Numerical, binary, and categorical features were examined separately to ensure that each variable was handled appropriately according to its data type. Initial screening included correlation checks, missingness assessment, and variance filtering to detect redundancy and identify features likely to contribute predictive value.

Although feature selection was later moved to occur post-imputation in the final pipeline, an initial round of preprocessing was still applied before imputation to remove features that were clearly uninformative or unsuitable for modeling. Specifically, we dropped identifier-like columns (`encounter_id`, `hospital_id`, `patient_id`, `icu_id`) as they carry no predictive signal, and removed numerical variables with more than 80% missingness or zero variance, as these would be difficult to impute and unlikely to add meaningful information. In total, 39 columns were removed during this initial filtering process.

```python
target = "hospital_death"
y = df[target]
X_unlabeled = df_test

# remove identifier columns
id_cols = ['encounter_id', 'hospital_id', 'patient_id', 'icu_id']
df.drop(columns=id_cols, inplace=True) # drop identifier columns

# identify numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
bin_cols = [col for col in num_cols if (df[col].nunique() == 2) & (col != target)] # do not include target
num_cols = num_cols.drop(list(bin_cols) + [target]) # remove binary and target columns from numerical features
cat_cols = df.select_dtypes(include=["object", "category"]).columns

# identify columns with >80% missing values
missing_rate = df[num_cols].isnull().mean()
high_missing_cols = missing_rate[missing_rate > 0.8].index.tolist()

# identify columns with zero standard deviation
df_temp = df[num_cols].fillna(df[num_cols].mean())
zero_std_cols = df_temp.columns[df_temp.std() == 0].tolist()

# remove highly missing and zero std columns from numerical features
filtered_num_cols = [c for c in num_cols if c not in high_missing_cols + zero_std_cols]
print(f"Features filtered out (missing > 80% or SD=0): {len(high_missing_cols) + len(zero_std_cols)}")
num_cols = filtered_num_cols

X = df[list(filtered_num_cols) + list(bin_cols) + list(cat_cols)]
X_unlabeled = df_test[list(filtered_num_cols) + list(bin_cols) + list(cat_cols)]
```

## 5.2 Imputation

Our feature set contains a mix of **numerical**, **categorical**, and **binary** variables, each requiring different handling during missing-value treatment. To determine the most suitable imputation strategy for each feature type, we conducted a systematic benchmarking experiment, comparing reconstruction accuracy under multiple imputation approaches. For **numerical predictors**, we evaluated four commonly used strategies - **mean**, **median**, **kNN**, and **MICE (IterativeImputer)** - and measured performance using RMSE under repeated random missingness masks. For **categorical variables**, missing entries were imputed using the **most frequent category**, as categorical distributions are often skewed and mode-based replacement preserves the natural class balance. For **binary features**, we applied **constant-value imputation (0)**, based on domain inspection suggesting that missing entries likely follow a MAR pattern and tend to represent the absence or non-severe presentation of a condition. This allowed us to retain interpretability while avoiding spurious intermediate values.

The full benchmarking workflow, implementation details, and validation metrics are available in a separate analysis notebook, `imputation.ipynb`. Across all numeric imputation methods tested, **MICE consistently achieved the lowest RMSE and provided the most accurate recovery of continuous values**, likely due to its ability to borrow information from correlated predictors. Based on these results, **MICE is selected as our default imputation method for continuous features moving forward**. Mean and median imputers remain useful as lightweight baselines or for sensitivity comparison, while kNN offers a practical compromise between performance and computational cost. Future extensions will explore MAR/MNAR simulations, measure the downstream effect on model performance, and evaluate whether lower RMSE correlates with improved prediction accuracy end-to-end.

```python
num_pipe = Pipeline([
    ("impute", IterativeImputer(max_iter=20, random_state=42, sample_posterior=True)),
    ("scale", StandardScaler())
])

# BINARY = mode impute (no scaling)
bin_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="most_frequent"))
])

# CATEGORICAL = mode impute → OneHot encode
cat_pipe = Pipeline([
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("encode", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer([
    ("num", num_pipe, num_cols),
    ("bin", bin_pipe, bin_cols),
    ("cat", cat_pipe, cat_cols)
], remainder="drop")

# preprocess
X_processed = preprocess.fit_transform(X)
feature_names = preprocess.get_feature_names_out() # get feature names after preprocessing
```

## 5.4 K-means Clustering
To further explore potential patient subgroups, we applied KMeans clustering to the PCA-reduced data. Because the first 46 principal components captured approximately 90% of the total variance, we performed KMeans in the 46-dimensional PCA space.

To evaluate cluster quality, we computed Silhouette scores on a 5000 sample subset of the PCA-46 representation. The highest Silhouette score occurred at k = 2 and consistently declined as k increased. This indicates that the dataset does not contain naturally well-separated high-order clusters, and increasing the number of clusters beyond three leads to over-partitioning rather than discovering meaningfully distinct groups.

```python
pca_53 = PCA(n_components=53)
X_pca_53 = pca_53.fit_transform(X_processed)

sample_idx = np.random.choice(len(X_pca_53), size=5000, replace=False)
X_sample = X_pca_53[sample_idx]

sil_scores = {}
cluster_assignments = {}

for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters_53 = kmeans.fit_predict(X_pca_53)
    cluster_assignments[k] = clusters_53
    
    clusters_sample = kmeans.predict(X_sample)
    sil = silhouette_score(X_sample, clusters_sample)
    sil_scores[k] = sil

sil_scores
```

We then visualized the clusters in the 2D PCA space using the first two principal components. Although the clusters appear cleanly separated in the plot, PC1 and PC2 together explain only about 20% of the total variance, so this apparent separation does not reflect true geometric separability in the original high-dimensional space.

```python
k = 2
clusters_k = cluster_assignments[k]
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_2[:, 0], X_pca_2[:, 1], c=clusters_k, cmap="tab10", alpha=0.6)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("KMeans Clusters in PCA-2 Space")
plt.show()
```

Calculating the mortality rate for each cluster, we found a clear difference between the two clusters, indicating possible underlying risk stratification in the patient population.

```python
cluster_series = pd.Series(clusters_k, index=y.index) 
mortality_by_cluster = y.groupby(cluster_series).mean()

print("Cluster-level mortality rates:")
print(mortality_by_cluster)
```

Our baseline model uses a multiple logistic regression classifier trained on preprocessed predictors to predict hospital_death. Overall, the model demonstrates strong discrimination ability, but also reveals clear limitations that motivate the need for more advanced modeling.


### 6.1.1 Model Architecture
We use Multiple Logistic Regression presented in the class:

$$
\ln\left( \frac{P(Y = 1 \mid X)}{1 - P(Y = 1 \mid X)} \right)
= \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p .
$$

Define the linear predictor:

$$
z = \beta_0 + \beta_1 X_1 + \cdots + \beta_p X_p .
$$

$$
P(Y = 1 \mid X) = \sigma(z)
= \frac{1}{1 + e^{-z}} .
$$

```python
# modeling pipeline 
logreg = Pipeline([
   ("preprocess", preprocess),
   ("model", LogisticRegression(max_iter=1000))
])
```

### 6.1.2 Cross-Validation Performance

Using 10-fold stratified cross-validation: **Mean Accuracy: ~0.925**, **Mean ROC AUC: ~0.878**

These values are consistent across folds, suggesting a stable model that generalizes reasonably well. The AUC close to 0.88 indicates that the model can reliably distinguish between survivors and non-survivors.

```python
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

cv_results = cross_validate(
   logreg,
   X,
   y,
   cv=cv,
   scoring=["accuracy", "roc_auc"],
   return_train_score=False
)

print("Accuracy:", cv_results["test_accuracy"].mean())
print("ROC AUC:", cv_results["test_roc_auc"].mean())
```

## 6.2  L1-regularized logistic regression (LASSO)
For another option, we proposed to use logistic regression model using an L1 penalty (LASSO) that predicts `hospital_death` using predictors preprocessed from above. 

L1 regularization forces many coefficients to zero, effectively performing automatic feature selection during model training. This allows us to identify the most predictive clinical variables while avoiding overfitting in a high-dimensional dataset.

### 6.2.1 Model Architecture

The L1-penalized objective function (LASSO logistic regression) is:

$$
\min_{\beta} \left[
    -\sum_{i=1}^{n}
        \left(
            y_i \log p_i
            + (1 - y_i)\log(1 - p_i)
        \right)
    \;+\;
    \lambda \sum_{j=1}^{p} |\beta_j|
\right],
$$

where $\lambda$ controls the strength of the L1 penalty.  
Larger $\lambda$ values produce a **sparser** model by shrinking more coefficients to zero.

We used processed data that already gone through processes and imputations after cutoff. Using cross-validated tuning over a wide range of regularization strengths, the model selected a relatively small C = 0.1389, indicating that substantial regularization was beneficial for this dataset. This is consistent with the high dimensionality and multicollinearity typical of ICU clinical data, where many predictors are redundant or weakly informative.

```python
# Train/validation split
X_processed = X_processed_df # use imputed dataset after cutoff
y = df["hospital_death"]
features = X_processed.columns

X_train, X_val, y_train, y_val = train_test_split(
   X_processed,
   y,
   test_size=0.3,
   stratify=y,
   random_state=42
)

Cs = np.logspace(-2, 2, 8)

logreg_l1 = LogisticRegressionCV(
   penalty="l1",
   solver="saga",
   Cs=Cs,
   cv=3,
   scoring="roc_auc",
   class_weight="balanced",
   max_iter=5000,
   n_jobs=-1,
   random_state=42
)
print("Fitting L1-regularized model...")
logreg_l1.fit(X_train, y_train)

print("Chosen C:", logreg_l1.C_[0])
```

### 6.2.2 Validation Performance

On the validation set, the L1 model achieves: Accuracy: 0.803, ROC AUC: 0.881

The AUC is nearly identical to the baseline model’s AUC (~0.88), indicating that the model’s ability to discriminate between survivors and non-survivors is preserved despite aggressive coefficient shrinkage. However, overall accuracy decreases compared to the baseline because the L1 model is now more willing to predict the minority class (death), which increases false positives but improves sensitivity to high-risk cases. This behavior is aligned with the goal of identifying patients at elevated mortality risk.

## 6.3 Ridge-regularized logistic regression (L2 / $\text{R}^\text{2}$ penalty)
For another regularization option, we fit a ridge (L2) logistic regression that predicts `hospital_death` using the same preprocessed predictors.
The squared L2 penalty discourages large weights, shrinking correlated coefficients together so that we stabilize estimates without forcing sparsity.

### 6.3.1 Model Architecture
We again rely on multiple logistic regression:

$$
\ln\left( \frac{P(Y = 1 \mid X)}{1 - P(Y = 1 \mid X)} \right)
= \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p .
$$

with linear predictor

$$
z = \beta_0 + \beta_1 X_1 + \cdots + \beta_p X_p ,
$$

and probability

$$
P(Y = 1 \mid X) = \sigma(z) = \frac{1}{1 + e^{-z}} .
$$

The L2-penalized objective (ridge logistic regression) augments the negative log-likelihood with the squared coefficient norm:

$$
\min_{\beta} \left[
    -\sum_{i=1}^{n} \Big( y_i \log p_i + (1 - y_i)\log(1 - p_i) \Big)
    + \lambda \sum_{j=1}^{p} \beta_j^{2}
\right],
$$

where $\lambda$ controls the amount of shrinkage. Larger $\lambda$ values push correlated coefficients toward one another, limiting variance while preserving all predictors.


```python
Cs_l2 = np.logspace(-3, 2, 10)
logreg_l2 = LogisticRegressionCV(
   penalty="l2",
   solver="lbfgs",
   Cs=Cs_l2,
   cv=5,
   scoring="roc_auc",
   class_weight="balanced",
   max_iter=5000,
   n_jobs=-1,
   random_state=42
)

print("Fitting L2-regularized logistic regression (ridge)...")
logreg_l2.fit(X_train, y_train) # type: ignore
print("Chosen C:", logreg_l2.C_[0])

y_val_proba_l2 = logreg_l2.predict_proba(X_val)[:, 1]
y_val_pred_l2 = logreg_l2.predict(X_val)

ridge_metrics = {
   "Accuracy": accuracy_score(y_val, y_val_pred_l2),
   "Precision": precision_score(y_val, y_val_pred_l2, zero_division=0),
   "Recall": recall_score(y_val, y_val_pred_l2),
   "F1": f1_score(y_val, y_val_pred_l2),
   "ROC AUC": roc_auc_score(y_val, y_val_proba_l2),
   "PR AUC": average_precision_score(y_val, y_val_proba_l2)
}

print("Validation metrics:")
for metric, value in ridge_metrics.items():
   print(f"  {metric}: {value:.3f}")

print("Classification report (ridge):")
print(classification_report(y_val, y_val_pred_l2, digits=3))
```

## 6.4 Elastic-net logistic regression
Elastic-net regularization combines the sparsity of L1 with the grouped shrinkage of L2, enabling us to adaptively balance feature selection and coefficient stability when modeling `hospital_death`.

### 6.4.1 Model Architecture
We reuse the same multiple logistic regression form:

$$
\ln\left( \frac{P(Y = 1 \mid X)}{1 - P(Y = 1 \mid X)} \right) = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p ,
$$

with linear predictor $z = \beta_0 + \beta_1 X_1 + \cdots + \beta_p X_p$ and probability $P(Y = 1 \mid X) = \sigma(z) = 1/(1 + e^{-z})$.

The elastic-net objective augments the negative log-likelihood with a convex combination of L1 and L2 penalties:

$$
\min_{\beta} \left[ -\sum_{i=1}^{n} \Big( y_i \log p_i + (1 - y_i)\log(1 - p_i) \Big) + \lambda \left(\alpha \sum_{j=1}^{p} |\beta_j| + \frac{1 - \alpha}{2} \sum_{j=1}^{p} \beta_j^{2}\right) \right],
$$

where $\alpha \in [0, 1]$ controls the mix between L1 and L2 penalties. Smaller $\alpha$ values behave more like ridge, while larger values resemble LASSO. Jointly tuning $\alpha$ and the overall penalty strength $\lambda$ lets us explore a continuum of regularized logistic models.


```python
enet_param_grid = {
   "C": np.logspace(-3, 1, 6),
   "l1_ratio": [0.1, 0.3, 0.5, 0.7, 0.9]
}

elastic_net = LogisticRegression(
   penalty="elasticnet",
   solver="saga",
   class_weight="balanced",
   max_iter=5000,
   random_state=42
)

grid_enet = GridSearchCV(
   estimator=elastic_net,
   param_grid=enet_param_grid,
   scoring="roc_auc",
   cv=3,
   n_jobs=-1,
   refit=True
)

print("Tuning elastic-net logistic regression...")
grid_enet.fit(X_train, y_train)
print("Best params:", grid_enet.best_params_)

logreg_enet = grid_enet.best_estimator_
y_val_proba_enet = logreg_enet.predict_proba(X_val)[:, 1]
y_val_pred_enet = logreg_enet.predict(X_val)

elastic_metrics = {
   "Accuracy": accuracy_score(y_val, y_val_pred_enet),
   "Precision": precision_score(y_val, y_val_pred_enet, zero_division=0),
   "Recall": recall_score(y_val, y_val_pred_enet),
   "F1": f1_score(y_val, y_val_pred_enet),
   "ROC AUC": roc_auc_score(y_val, y_val_proba_enet),
   "PR AUC": average_precision_score(y_val, y_val_proba_enet)
}

print("Validation metrics (elastic net):")
for metric, value in elastic_metrics.items():
   print(f"  {metric}: {value:.3f}")

print("Classification report (elastic net):")
print(classification_report(y_val, y_val_pred_enet, digits=3))
```

# 7. Final Model Pipeline

Our final model extends the baseline logistic regression framework by incorporating a stacked ensemble architecture that integrates five diverse classifiers: decision tree, random forest, rotation forest, calibrated linear SVM, and ridge logistic regression. Each base learner captures different structural patterns in the feature space—tree-based methods model nonlinear interactions, SVM emphasizes margin separation, and ridge logistic regression offers linear stability. By leveraging this model diversity, our ensemble aims to reduce variance and improve overall predictive performance.

To construct the stacked model, we first train all base learners on the training set and generate cross-validated out-of-fold predictions. Using out-of-fold predictions prevents information leakage and provides the meta-learner with an unbiased representation of each model’s behavior. We use predicted class probabilities (rather than hard labels) as inputs to a second-stage logistic regression meta-model. This top-layer model is regularized, which helps control overfitting introduced by stacking and preserves interpretability through feature shrinkage.

The final inference pipeline operates in two stages. First, a new observation is passed through the five base learners to obtain a vector of predicted probabilities. Second, these outputs are concatenated and fed into the meta-logistic regression, which produces the final mortality probability. This configuration retains the simplicity and interpretability of logistic regression at the top level, while benefiting from the nonlinear representational capacity of tree-based models and the margin-based discriminability of SVM. Overall, the stacked ensemble yields a more flexible, robust, and accurate classifier than the baseline logistic regression alone.


### 7.1.3 Rotation Forest

Rotation Forest applies PCA-based rotations to random feature subsets before fitting each tree, creating diverse decision boundaries while keeping the full information content of the data. To keep computation tractable on our high-dimensional ICU table, we implement a lightweight variant that (1) uses small feature blocks with randomized SVD (no full covariance inversion), (2) limits tree depth and minimum leaf size, and (3) keeps the number of estimators modest. This preserves the core idea of rotated feature spaces while reducing both memory and training time relative to a full RotationForest implementation.


```python
class FastRotationForest(BaseEstimator, ClassifierMixin):
    """Rotation Forest with randomized SVD blocks to reduce compute."""

    def __init__(self, n_estimators=24, subset_size=12, max_depth=8, min_samples_leaf=3, random_state=None):
        self.n_estimators = n_estimators
        self.subset_size = subset_size
        self.max_depth = max_depth
        self.min_samples_leaf = min_samples_leaf
        self.random_state = random_state

    def _build_rotations(self, X, rng):
        n_features = X.shape[1]
        perm = rng.permutation(n_features)
        subsets = [perm[i:i + self.subset_size] for i in range(0, n_features, self.subset_size)]
        rotations = []
        rotated_blocks = []
        for subset in subsets:
            svd = TruncatedSVD(
                n_components=len(subset),
               random_state=rng.randint(0, 1_000_000)
            )
            rotated_blocks.append(svd.fit_transform(X[:, subset]))
            rotations.append((subset, svd))
        return rotations, rotated_blocks

    def fit(self, X, y):
        X, y = check_X_y(X, y, accept_sparse=False)
        X = X.astype(np.float32, copy=False)
        rng = check_random_state(self.random_state)
        self.estimators_ = []
        self.rotations_ = []
        self.n_features_in_ = X.shape[1]

        for _ in range(self.n_estimators):
            rotations, blocks = self._build_rotations(X, rng)
            X_rot = np.concatenate(blocks, axis=1)
            tree = DecisionTreeClassifier(
                max_depth=self.max_depth,
                min_samples_leaf=self.min_samples_leaf,
                class_weight="balanced",
                random_state=rng.randint(0, 1_000_000)
            )
            tree.fit(X_rot, y)
            self.estimators_.append(tree)
            self.rotations_.append(rotations)

        self.classes_ = self.estimators_[0].classes_
        return self

    def _transform(self, X, rotations):
        blocks = [svd.transform(X[:, subset]) for subset, svd in rotations]
        return np.concatenate(blocks, axis=1)

    def predict_proba(self, X):
        check_is_fitted(self, ["estimators_", "rotations_"])
        X = check_array(X, accept_sparse=False).astype(np.float32, copy=False)
        proba = np.zeros((X.shape[0], len(self.classes_)), dtype=float)
        for tree, rotations in zip(self.estimators_, self.rotations_):
            X_rot = self._transform(X, rotations)
            proba += tree.predict_proba(X_rot)
        return proba / len(self.estimators_)

    def predict(self, X):
        probs = self.predict_proba(X)
        return self.classes_[np.argmax(probs, axis=1)]

fast_rot_forest = FastRotationForest(
    n_estimators=24,
    subset_size=12,
    max_depth=8,
    min_samples_leaf=3,
    random_state=42
)

fast_rot_forest.fit(X_train, y_train)
```

### 7.1.4 Linear SVM (calibrated)

A linear SVM provides a high-margin linear separator for the sparse, high-dimensional feature space. To keep computation reasonable, we tune the penalty parameter on a 15k-sample subset with 3-fold CV and then apply Platt scaling (sigmoid calibration) to recover class probabilities needed for stacking.


```python
# randomly sample up to 15k rows for tuning
svm_tune_idx = np.random.default_rng(42).choice(
    len(X_train), size=min(len(X_train), 15000), replace=False
)

# convert to DataFrame for easier indexing
X_train_df = pd.DataFrame(X_train, columns=feature_names)
y_train_df = pd.DataFrame(y_train, columns=["hospital_death"])

# select sampled subset
X_tune_svm = X_train_df.iloc[svm_tune_idx]
y_tune_svm = y_train_df.iloc[svm_tune_idx]

print(type(y_train), type(X_train))
print(X_train.shape, y_train.shape)

svm_candidates = [0.25, 0.5, 1.0]
svm_auc_scores = {}
svm_cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

for C in svm_candidates:
    linear_svm = LinearSVC(
        C=C,
        penalty="l2",
        dual=False,
        class_weight="balanced",
        random_state=42
    )
    scores = cross_val_score(
        linear_svm,
        X_tune_svm,
        y_tune_svm,
        cv=svm_cv,
        scoring="roc_auc",
        n_jobs=-1
    )
    svm_auc_scores[C] = scores.mean()

best_svm_C = max(svm_auc_scores, key=svm_auc_scores.get)
print("Linear SVM ROC-AUC (tuning subset):", svm_auc_scores)
print("Chosen C:", best_svm_C)

svm_linear = LinearSVC(
    C=best_svm_C,
    penalty="l2",
    dual=False,
    class_weight="balanced",
    random_state=42
)
svm_calibrated = CalibratedClassifierCV(
    estimator=svm_linear,
    method="sigmoid",
    cv=svm_cv,
    n_jobs=-1
)
svm_calibrated.fit(X_train, y_train)

y_val_proba_svm = svm_calibrated.predict_proba(X_val)[:, 1]
y_val_pred_svm = svm_calibrated.predict(X_val)

svm_metrics = {
    "Accuracy": accuracy_score(y_val, y_val_pred_svm),
    "Precision": precision_score(y_val, y_val_pred_svm, zero_division=0),
    "Recall": recall_score(y_val, y_val_pred_svm),
    "F1": f1_score(y_val, y_val_pred_svm),
    "ROC AUC": roc_auc_score(y_val, y_val_proba_svm),
    "PR AUC": average_precision_score(y_val, y_val_proba_svm)
}
print("Validation metrics (linear SVM):")
for metric, value in svm_metrics.items():
    print(f"  {metric}: {value:.4f}")
```


### 7.1.5 Ridge Logistic Regression (stacking base)

Among the logistic models above, the ridge-regularized variant delivered the best balance of ROC/PR AUC and calibration (Section 6.3 tuned $C \approx 0.046$). We reuse that setting, refit on the current 80/20 split, and keep it as a linear, well-calibrated base learner for stacking.


```python
ridge_base = LogisticRegression(
    penalty="l2",
    C=0.05,
    solver="lbfgs",
    class_weight="balanced",
    max_iter=4000,
    n_jobs=-1,
    random_state=42
)

ridge_base.fit(X_train, y_train)

y_val_proba_ridge_base = ridge_base.predict_proba(X_val)[:, 1]
y_val_pred_ridge_base = ridge_base.predict(X_val)

ridge_base_metrics = {
    "Accuracy": accuracy_score(y_val, y_val_pred_ridge_base),
    "Precision": precision_score(y_val, y_val_pred_ridge_base, zero_division=0),
    "Recall": recall_score(y_val, y_val_pred_ridge_base),
    "F1": f1_score(y_val, y_val_pred_ridge_base),
    "ROC AUC": roc_auc_score(y_val, y_val_proba_ridge_base),
    "PR AUC": average_precision_score(y_val, y_val_proba_ridge_base)
}

print("Ridge logistic (stacking base) metrics:")
for metric, value in ridge_base_metrics.items():
    print(f" {metric}: {value:.4f}")
```

## 7.2 Stacking the Base Learners

We combine the five base learners (decision tree, random forest, rotation forest, calibrated linear SVM, and ridge logistic regression) using a `StackingClassifier`. The meta-learner is a class-weighted logistic regression trained on out-of-fold predicted probabilities from each base model. To keep runtime reasonable we use 3-fold stacking CV and `n_jobs=-1`.


```python
stack_estimators = [
    ("decision_tree", dt_best),
    ("random_forest", rf_pipeline),
    ("rotation_forest", fast_rot_forest),
    ("linear_svm", svm_calibrated),
    ("ridge_logistic", ridge_base),
]

meta_learner = LogisticRegression(
    penalty="l2",
    C=1.0,
    solver="lbfgs",
    class_weight="balanced",
    max_iter=3000,
    n_jobs=-1,
    random_state=42
)

stacking_clf = StackingClassifier(
    estimators=stack_estimators,
    final_estimator=meta_learner,
    stack_method="predict_proba",
    cv=3,
    n_jobs=-1,
    passthrough=False
)

stacking_clf.fit(X_train, y_train)

stack_proba = stacking_clf.predict_proba(X_val)[:, 1]
stack_pred = stacking_clf.predict(X_val)

stack_metrics = {
    "Accuracy": accuracy_score(y_val, stack_pred),
    "Precision": precision_score(y_val, stack_pred, zero_division=0),
    "Recall": recall_score(y_val, stack_pred),
    "F1": f1_score(y_val, stack_pred),
    "ROC AUC": roc_auc_score(y_val, stack_proba),
    "PR AUC": average_precision_score(y_val, stack_proba)
}

print("Stacked ensemble metrics:")
for metric, value in stack_metrics.items():
    print(f"  {metric}: {value:.4f}")

stack_cm = confusion_matrix(y_val, stack_pred)
plt.figure(figsize=(6,4))
sns.heatmap(stack_cm, annot=True, fmt='d', cmap='Oranges',
            xticklabels=['Survived', 'Died'], yticklabels=['Survived', 'Died'])
plt.title(f"Stacking Confusion Matrix\\nROC AUC: {stack_metrics['ROC AUC']:.3f}")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```


## 8. Review and Sensitivity Analysis

We summarize validation metrics across all base learners and the stacked model, then probe how the stacked predictions behave under different decision thresholds and subsampled validation folds (a lightweight stability check without retraining).


```python
summary_rows = [
    {"Model": "Decision Tree", **dt_metrics},
    {"Model": "Random Forest", **rf_metrics},
    {"Model": "Rotation Forest", **rot_metrics},
    {"Model": "Linear SVM", **svm_metrics},
    {"Model": "Ridge Logistic", **ridge_base_metrics},
    {"Model": "Stacked Ensemble", **stack_metrics},
]

summary_df = pd.DataFrame(summary_rows)
summary_df = summary_df.set_index("Model").loc[
    ["Decision Tree", "Random Forest", "Rotation Forest", "Linear SVM", "Ridge Logistic", "Stacked Ensemble"]
]

print("Validation metrics comparison (higher is better):")
display(summary_df.sort_values("ROC AUC", ascending=False))

thresholds = [0.2, 0.3, 0.4, 0.5]
threshold_rows = []
for t in thresholds:
    preds = (stack_proba >= t).astype(int)
    threshold_rows.append({
        "Threshold": t,
        "Precision": precision_score(y_val, preds, zero_division=0),
        "Recall": recall_score(y_val, preds),
        "Specificity": recall_score(y_val, preds, pos_label=0),
        "F1": f1_score(y_val, preds),
        "Accuracy": accuracy_score(y_val, preds)
    })

threshold_df = pd.DataFrame(threshold_rows)
print("\nStacked model threshold sweep:")
display(threshold_df)

rng = np.random.default_rng(42)
bootstrap_rows = []
for seed in range(3):
    idx = rng.choice(len(y_val), size=int(0.7 * len(y_val)), replace=False)
    y_boot = y_val.iloc[idx]
    proba_boot = stack_proba[idx]
    preds_boot = (proba_boot >= 0.3).astype(int)
    bootstrap_rows.append({
        "Seed": seed,
        "ROC AUC": roc_auc_score(y_boot, proba_boot),
        "PR AUC": average_precision_score(y_boot, proba_boot),
        "F1 @0.3": f1_score(y_boot, preds_boot),
        "Accuracy @0.3": accuracy_score(y_boot, preds_boot)
    })

bootstrap_df = pd.DataFrame(bootstrap_rows)
print("\nBootstrap sensitivity on validation (70% subsamples, threshold=0.3):")
display(bootstrap_df)
```
